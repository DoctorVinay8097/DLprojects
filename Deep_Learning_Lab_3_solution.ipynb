## **Implementation of forward propagation of a 3-Layer Neural Network**

image.png

* Even though we have readily available packages like **Keras and Tensorflow** for forward and backward propagation, its **essential** for a student to know how they are **implemented at the backend**.
* So in this lab, we are going to code the **forward propagation of a 3-Layer Neural network**.
* The present practical is for a **binary classification**, so the number of neurons in the output layer will be **one**.

```python
import numpy as np
```

### **Step 1:** Make a pratice of dividing the problem into small parts and code them as individual blocks i.e. functions.
* Define the functions for sigmoid and relu activation function

```python
# Define ReLU activation function
def relu(x):
    return np.maximum(0, x)

# Define sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

### **Step 2:** Create some random input data (X) with ground truth (y) labels.

```python
# Define input data
input_data = -10 + np.random.rand(5, 3)*20  # 5 examples, each with 3 features
print(input_data)
```

```python
# Generate random binary y labels
y_labels = np.random.randint(2, size=(5, 1))
print(y_labels)
```

### **Step 3:** Initialize the weights and biases with some random values for each layer.
* Remember the number of weights and biases required for the **hidden layer** will depend on the **input data**.
* For the **output layer**, the number of weights and biases required will depend on the **hidden layer**.

```python
# Define weights and biases for the hidden layer
hidden_weights = np.random.rand(10, 3)  # 3 input features, 10 hidden neurons
hidden_biases = np.random.rand(10, 1)    # Biases for each hidden neuron

print("Initial hidden weights: \n", hidden_weights)
```

```python
print("Initial hidden biases: \n", hidden_biases)
```

```python
# Define weights and bias for the output layer
output_weights = np.random.rand(1, 10)  # 10 hidden neurons, 1 output neuron
output_bias = np.random.rand(1, 1)       # Bias for the output neuron
print("Initial output weights: \n", output_weights)
```

```python
print("Initial output bias", output_bias)
```

### **Step 4:** Forward propagate from input layer to the hidden layer

```python
# Forward propagation
# Input layer to hidden layer
hidden_weighted_sum = np.dot(hidden_weights, input_data.T)
```

```python
hidden_weighted_sum = hidden_weighted_sum + hidden_biases
print("hidden_weighted_sum_with bias: \n", hidden_weighted_sum)
```

```python
hidden_activations = relu(hidden_weighted_sum)
print(hidden_activations)
```

### **Step 5:** Forward propagate from hidden layer to the output layer

```python
# Hidden layer to output layer
output_weighted_sum = np.dot(output_weights, hidden_activations)
print("output_weighted_sum_without_bias", output_weighted_sum)
```

```python
output_weighted_sum = output_weighted_sum + output_bias
print("output_weighted_sum_with_bias", output_weighted_sum)
```

```python
output = sigmoid(output_weighted_sum)
print("y_prediction: \n", output.T)
```

```python
print(y_labels)
```

### **Step 6:** Find the Loss for each example and compute the final cost J for the given dataset.

```python
# Define binary cross-entropy loss
def binary_cross_entropy(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # Clip to prevent log(0)
    return - (y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))
```

```python
# Calculate losses for each instance using binary cross-entropy loss
losses = binary_cross_entropy(y_labels, output)

# Compute the final cost J (average loss)
J = np.mean(losses)
```

```python
print("Final Cost J: ", J)
```
