```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def kernel(vertical = True):
  if vertical == True:
    k = np.array([[1, 0, -1],
                  [1, 0, -1],
                  [1, 0, -1]])
  else:
    k = np.array([[1, 1, 1],
                  [0, 0, 0],
                  [-1, -1, -1]])

  return k
```

```python
def sobel_kernel(vertical = True):
  if vertical == True:
    k = np.array([[1, 0, -1],
                  [2, 0, -2],
                  [1, 0, -1]])
  else:
    k = np.array([[1, 2, 1],
                  [0, 0, 0],
                  [-1, -2, -1]])

  return k
```

```python
def convol(in_img, hori = False):
  # Perform convolution using NumPy
  final_shape = (in_img.shape[0] - 3) + 1
  output_image = np.zeros((final_shape, final_shape), dtype=np.int32)

  for i in range(final_shape):
    for j in range(final_shape):
        output_image[i, j] = np.sum(in_img[i:i+3, j:j+3] * kernel(vertical = not hori))

  return output_image
```

```python
def plot_images(in_img, out_img):
  plt.imshow(in_img, cmap='gray')
  plt.axis('off')  # Turn off axis labels
  plt.show()
  plt.imshow(out_img, cmap='gray')
  plt.axis('off')  # Turn off axis labels
  plt.show()
```

```python

from skimage import data

# Load the Camera Man grayscale image from scikit-image
image = data.camera()


plot_images(image, convol(image, True))
```
